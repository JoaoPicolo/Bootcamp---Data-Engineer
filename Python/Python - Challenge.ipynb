```python
usersDF = pd.read_csv("/content/drive/MyDrive/users.csv")
usersDF.head()
```

```python
# Number of unique countries and continents
print(len(usersDF["Country or area"].unique()))
```

```python
# Class defined by the challenge description
class MapWithContinent:
  continentMap = {}

  def mapping(self, df: pd.DataFrame, countryCol: str, continentCol: str):
    countries = df[countryCol].tolist()
    continents = df[continentCol].tolist()

    for idx, country in enumerate(countries):
      self.continentMap[country] = continents[idx]

  def getContinent(self, country: str):
    try:
        return self.continentMap[country]
    except:
        print(f"There's no country called { country } in our base")
        return None
```

```python
countriesMap = MapWithContinent()
countriesMap.mapping(usersDF, "Country or area", "Region")
countriesMap.getContinent("Brazil")
```

```python
speedsDF = pd.read_csv("/content/drive/MyDrive/avgspeed.csv", header=0, names=["Country", "Avg Speed"])
speedsDF.head()
```

```python
# Adds column Continent to speeds dataframe
speedsDF['Continent'] = speedsDF.apply(lambda row: countriesMap.getContinent(row["Country"]), axis=1)
speedsDF.head()

print(speedsDF["Continent"].isna().sum()) # Number on null continents
print(speedsDF["Continent"].count())      # Number of valid continents
print(speedsDF["Avg Speed"].mean())       # Gets the avg speed acroos all countries
print(speedsDF.groupby(["Continent"])["Avg Speed"].mean()) # Gets the avg speed by continet
```

```python
# Reads new instance of users
newUsersDF = pd.read_csv("/content/drive/MyDrive/users.csv")
print(newUsersDF["Internet users"].isna().sum())
print(newUsersDF["Population"].isna().sum())

# Drop null values
newUsersDF.dropna(inplace=True)
print(newUsersDF["Internet users"].isna().sum())
print(newUsersDF["Population"].isna().sum())
```

```python
# Converts column from str to int
newUsersDF["Internet users"] = newUsersDF["Internet users"].apply(lambda x: x.replace(',', ''))
newUsersDF["Internet users"] = newUsersDF["Internet users"].astype(int)

# Converts column from str to int
newUsersDF["Population"] = newUsersDF["Population"].apply(lambda x: x.replace(',', ''))
newUsersDF["Population"] = newUsersDF["Population"].astype(int)

newUsersDF.head()
```

```python
# Get countries where more than 90% of the population has access to the internet
bigAccessDF = newUsersDF[newUsersDF["Internet users"] > (newUsersDF["Population"] * 0.9)]
bigAccessDF.count()
```

```python
# Get countries where less than 10% of the population has access to the internet
smallAccessDF = newUsersDF[newUsersDF["Internet users"] < (newUsersDF["Population"] * 0.1)]
smallAccessDF.count()
```
